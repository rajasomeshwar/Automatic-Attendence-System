```python
# Import the required modules
import mysql.connector
from PIL import Image
import io

# For security reasons, never expose your password
password = 'root'

# Create a connection
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password=password,
    database="world"  # Name of the database
)

# Create a cursor object
cursor = mydb.cursor()

# Prepare the query
query = 'SELECT PICTURE FROM studentD WHERE ID=105'

# Execute the query to get the file
cursor.execute(query)

data = cursor.fetchone()

if data is not None:
    # The returned data will be a binary object
    image_data = data[0]

    # Convert the bytes into a PIL image
    image = Image.open(io.BytesIO(image_data))

    # Display the image
    image.show()
else:
    print("No image found for the given ID.")

# Close the cursor and connection
cursor.close()
mydb.close()
```

```python
import cv2
import numpy as np
from mtcnn import MTCNN
from keras.models import load_model
from sklearn.preprocessing import Normalizer
from sklearn.svm import SVC
import os
import pickle

# Load the FaceNet model
facenet_model = load_model('facenet_keras.h5')
detector = MTCNN()

# L2 Normalization
in_encoder = Normalizer(norm='l2')

def extract_face(image, detector):
    results = detector.detect_faces(image)
    if len(results) > 0:
        x, y, width, height = results[0]['box']
        face = image[y:y+height, x:x+width]
        return face
    return None

def get_embedding(model, face_pixels):
    face_pixels = face_pixels.astype('float32')
    mean, std = face_pixels.mean(), face_pixels.std()
    face_pixels = (face_pixels - mean) / std
    samples = np.expand_dims(face_pixels, axis=0)
    yhat = model.predict(samples)
    return yhat[0]

def load_dataset(data_dir):
    faces, ids = [], []
    for filename in os.listdir(data_dir):
        path = os.path.join(data_dir, filename)
        image = cv2.imread(path)
        face = extract_face(image, detector)
        if face is not None:
            face = cv2.resize(face, (160, 160))
            embedding = get_embedding(facenet_model, face)
            faces.append(embedding)
            ids.append(int(filename.split(".")[1]))
    return np.asarray(faces), np.asarray(ids)

def train_classifier(data_dir):
    faces, ids = load_dataset(data_dir)
    in_encoder.fit(faces)
    faces = in_encoder.transform(faces)
    
    # Train an SVM classifier
    classifier = SVC(kernel='linear', probability=True)
    classifier.fit(faces, ids)
    
    # Save the classifier
    with open('face_classifier.pkl', 'wb') as file:
        pickle.dump(classifier, file)

def recognize_face(image, detector, model, classifier):
    face = extract_face(image, detector)
    if face is not None:
        face = cv2.resize(face, (160, 160))
        embedding = get_embedding(model, face)
        embedding = in_encoder.transform([embedding])
        id = classifier.predict(embedding)
        return id[0]
    return None

# Train the model
train_classifier('data')

# Load the classifier
with open('face_classifier.pkl', 'rb') as file:
    face_classifier = pickle.load(file)

# Use the classifier in your real-time face recognition script
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    id = recognize_face(frame, detector, facenet_model, face_classifier)
    if id:
        print(f"Recognized ID: {id}")
    
    cv2.imshow("Face Recognition", frame)
    if cv2.waitKey(1) == 13:
        break

cap.release()
cv2.destroyAllWindows()
```
